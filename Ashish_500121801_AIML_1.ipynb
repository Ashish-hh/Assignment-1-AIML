```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Specify the path to the file in your Google Drive
file_path = '/content/drive/My Drive/Country_Consumption_TWH.csv'

# Load the CSV file
data = pd.read_csv(file_path)

# Basic Summary
print(data.info())

# Display the first 5 rows
print(data.head())

# Check for missing values
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)

# Quick visualization of energy consumption trends for selected countries
selected_countries = ['China', 'United States', 'India']
for country in selected_countries:
    plt.plot(data['Year'], data[country], label=country)

plt.xlabel('Year')
plt.ylabel('Energy Consumption (TWh)')
plt.title('Energy Consumption Trends')
plt.legend()
plt.grid(True)
plt.show()
```
